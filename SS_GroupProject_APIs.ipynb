```python
import pandas as pd
import numpy as np
import requests
import json
from pandas.io.json import json_normalize
```

```python
# URL for GET requests to retrieve global count data and all countries data
url1 = "https://coronavirus-19-api.herokuapp.com/all"
url2 = "https://coronavirus-19-api.herokuapp.com/countries"

# URL for GET requests to retrieve US States TestResults data
url3 = "https://covidtracking.com/api/states"
```

```python
response = requests.get(url1)
response.json()
```

```python
Global_df = json_normalize(response.json(), errors='ignore')
Global_df
```

```python
response = requests.get(url2)
response.json()
```

```python
Countries_df = json_normalize(response.json(), errors='ignore')
Countries_df
```

```python
Countries_df.iloc[:,[0,1,2,3,4,5,6,7]]
```

```python
response = requests.get(url3)
response.json()
```

```python
US_df = json_normalize(response.json(), errors='ignore')
US_df
```

```python
US_df.iloc[:,[0,1,8,22]]
```
