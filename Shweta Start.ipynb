```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from census import Census
from pprint import pprint
```

```python
globses = pd.read_csv("GLOB.csv", encoding = 'latin-1')
globses = globses[globses["year"]==2010]
globses = globses.reset_index(drop = True)
globses
```

```python
countries = globses["country"]
countries= list(countries)
countries.index("United States")
countries[142]='USA'
countries[142]
```

```python
# URL for GET requests to retrieve global count data
url1 = "https://coronavirus-19-api.herokuapp.com/all"
# URL for GET requests to retrieve all countries data
url2 = "https://coronavirus-19-api.herokuapp.com/countries"

# Retrieving global data, converting it into JSON and Printing the output of the JSON
response = requests.get(url2).json()
pprint(response)

```

```python

# set up lists to hold reponse info
countries_used = []
active = []
cases = []
critical = []
deaths = []
recovered = []
todayCases = []
todayDeaths = []
totalTest = []
# Loop through the list of countries and perform a request for data on each
for i in range(len(response)):
    if response[i]["country"] in countries:
        countries_used.append(response[i]["country"])
        active.append(response[i]["active"])
        cases.append(response[i]["cases"])
        critical.append(response[i]["critical"])
        deaths.append(response[i]["deaths"])
        recovered.append(response[i]["recovered"])
        todayCases.append(response[i]["todayCases"])
        todayDeaths.append(response[i]["todayDeaths"])
        totalTest.append(response[i]["totalTests"])
        
    else:
        print(response[i]["country"])
   
```

```python
print(len(countries_used))
print(len(active))
print(len(cases))
print(len(critical))
print(len(deaths))
print(len(recovered))
print(len(todayCases))
print(len(todayDeaths))
print(len(totalTest))
```

```python
df = pd.DataFrame({"Countries":countries_used,
                   "Active Cases":active,
                   "Total Cases":cases,
                   "Critical":critical,
                  "Deaths":deaths,
                  "Recovered":recovered,
                  "Cases Today":todayCases,
                  "Deaths Today":todayDeaths,
                  "Tested":totalTest})
df
```
